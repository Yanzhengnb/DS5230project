```python
def extract_first_book(file_path):
    book = epub.read_epub(file_path)
    content = []
    found_first_book = False
    
    for item in book.get_items_of_type(ebooklib.ITEM_DOCUMENT):
        soup = BeautifulSoup(item.get_content(), 'html.parser')
        
        # Check for a unique heading or title marking the first book
        if not found_first_book:
            headings = soup.find_all(['h1', 'h2'])  # Adjust tag types as needed
            for heading in headings:
                if "Book One" in heading.text or "The First Book" in heading.text:
                    found_first_book = True
                    break
        
        # Start collecting content for the first book
        if found_first_book:
            content.append(soup.get_text())
            # Optional: Add a stopping condition (e.g., "Book Two")
            if soup.find(['h1', 'h2'], string=lambda text: "Book Two" in text if text else False):
                break
    
    return "\n".join(content)
```

```python
def process_first_books(input_files, output_folder):
    os.makedirs(output_folder, exist_ok=True)
    
    for file_path in input_files:
        file_name = os.path.basename(file_path)
        print(f"Processing: {file_name}")
        
        # Extract the first book
        first_book_text = extract_first_book(file_path)
        
        # Save the processed text
        output_file = os.path.join(output_folder, f"{os.path.splitext(file_name)[0]}_first_book.txt")
        with open(output_file, 'w', encoding='utf-8') as f:
            f.write(first_book_text)
```

```python

# Example usage
input_files = ["Data/raw/The Wheel of Time Books 1-10.epub", "Data/raw/The Books of Earthsea.epub"]
output_folder = "Data/processed"
process_first_books(input_files, output_folder)
```

```python
import pdfplumber

def extract_first_book_from_pdf(pdf_path, book_title, next_book_title):
    content = []
    found_start = False
    
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text = page.extract_text()
            
            if text is None:
                continue  # Skip pages with no text
                
            # Check if the book's title is present
            if not found_start and book_title in text:
                found_start = True
            
            # Start collecting content once the title is found
            if found_start:
                content.append(text)
                
                # Optional: Stop if another book's title is encountered
                if next_book_title in text:
                    break
    
    # Join the collected content into a single string
    return "\n".join(content)
```

```python
import re

def clean_text(text):
    # Remove excessive newlines and spaces
    cleaned_text = re.sub(r'\n+', '\n', text)  # Remove excessive newlines
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text)  # Collapse multiple spaces into one
    
    # Add spaces between words that are concatenated (e.g., 'HelloWorld' -> 'Hello World')
    # This regular expression looks for lowercase followed by uppercase letters, a common case in concatenated words
    cleaned_text = re.sub(r'([a-z])([A-Z])', r'\1 \2', cleaned_text)  # Adds space between lowercase and uppercase letters

    return cleaned_text
```

```python
import pdfplumber
import re

def clean_text(text):
    # Remove excessive newlines and spaces
    cleaned_text = re.sub(r'\n+', '\n', text)  # Remove excessive newlines
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text)  # Collapse multiple spaces into one
    
    # Fix concatenated words (e.g., 'HelloWorld' -> 'Hello World')
    cleaned_text = re.sub(r'([a-z])([A-Z])', r'\1 \2', cleaned_text)
    
    return cleaned_text

def extract_first_book_from_pdf(pdf_path, book_title, next_book_title):
    content = []
    found_start = False
    
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            words = page.extract_words()
            
            # Rebuild the text from extracted words
            page_text = ' '.join([word['text'] for word in words])
            
            if page_text is None or page_text.strip() == "":
                continue  # Skip empty or invalid pages
            
            # Check if the book's title is present
            if not found_start and book_title in page_text:
                found_start = True
            
            # Start collecting content once the title is found
            if found_start:
                content.append(page_text)
                
                # Stop if the next book's title is encountered
                if next_book_title in page_text:
                    break
    
    # Join the collected content into a single string and clean it
    return "\n".join([clean_text(text) for text in content])

# Example usage
pdf_path = "Data/raw/The Fellowship of the Ring.pdf"
book_title = "The Fellowship of the Ring"
next_book_title = "The Two Towers"  # Adjust based on next book title

first_book_text = extract_first_book_from_pdf(pdf_path, book_title, next_book_title)

# Save the processed text to a file
output_file = "Data/processed/The_Fellowship_of_the_Ring.txt"
with open(output_file, 'w', encoding='utf-8') as f:
    f.write(first_book_text)

print(f"First book saved to: {output_file}")
```
